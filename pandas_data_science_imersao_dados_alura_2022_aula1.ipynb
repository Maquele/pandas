## **1 - Média da metragem para cada um dos bairros**

```python
media_metragem_bairros = dados.groupby(['Bairro']).mean().round(2)
media_metragem_bairros
```

```python
media_metragem_bairros.sample(10)
```

## **2 - Duas formas de selecionar os dados por bairro**

```python
bairros = dados.groupby('Bairro')
```

```python
bairros.describe()['Metragem'].head(10).round(2)
```

```python
bairros.mean().head().round(2)
```

## **3 - Explorando gráficos**

```python
bairros = dados.groupby('Bairro')
```

```python
m = bairros.max()
print(m)
```

```python
maiores_imoveis_por_bairro = m.sort_values(by='Metragem', ascending=False).head()
maiores_imoveis_por_bairro.head()
```

```python
maiores_imoveis_por_bairro.plot.bar()
```
